```python
#This is typical beautiful soup code to extract the table and 3 empty lists are made

import unicodedata
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
#print(soup.prettify())
My_table = soup.find('table',{'class':'wikitable sortable'})
A=[]
B=[]
C=[]
```

```python
# This code fills the content from html extracts in the lists one by one while making sure the content gets converted to string format as it in in unicode form. 
#Also extra things are stripped away from list C

for row in My_table.findAll('tr'):
    cells = row.findAll('td')
    if len(cells)==3: 
        A.append(str(cells[0].find(text=True)))
        B.append(str(cells[1].find(text=True)))
        C.append(str(cells[2].find(text=True).strip('\n')))

```

```python
#Converting the 3 lists we got in a dataframe 
import pandas as pd
df=pd.DataFrame(A,columns=['Postcode'])
df['Borough']= B
df['Neighbourhood']= C

#you can see the intial dataframe 
df
```

```python
# This code drops the rows of dataframe based on where Borough column has not assigned in it.  

for index, row in df.iterrows():
    if row['Borough'] == 'Not assigned':
        df.drop(index, inplace=True)

#you can see the tranformed dataframe here        
df
```

```python
# here i converted the postcode column items as indexes of the dataframe for further analysis.

df.index = df['Postcode'] 

#see the tranformed dataframe 
df
```

```python
# First I take out the unique postcodes values from the postcode column so that we can use it while doing further operations.

unique_postcodes = df['Postcode'].unique()
unique_postcodes
```

```python
# As given in problem statement if the postcode has multiple neighbourhoods we need to integrate them in one row. 
# I converted the the row value i get into pandas series as it was convenient when updating the row values
# sorry for the multiple print statements as it was needed while debugging 

for x in range(len(unique_postcodes)):
    current_postcode = unique_postcodes[x]
    print("1st")
    current_row_col_value = df.loc[current_postcode]['Neighbourhood']
    print("2st")
    series_convert = pd.Series(current_row_col_value)
    print("3st")
    length_of_series = len(series_convert)
    print("4st")
    if length_of_series == 1:
        df.loc[current_postcode,'Neighbourhood'] = df.loc[current_postcode,'Neighbourhood']
        print("5st")
    elif length_of_series == 2:
        df.loc[current_postcode,'Neighbourhood'] = str(current_row_col_value.iloc[0] +","+ current_row_col_value.iloc[1])
        print("6st")
    elif length_of_series == 3:
        df.loc[current_postcode,'Neighbourhood'] = str(current_row_col_value.iloc[0] +","+ current_row_col_value.iloc[1]+","+ current_row_col_value.iloc[2])
        print("7st")
    elif length_of_series == 4:
        df.loc[current_postcode,'Neighbourhood'] = str(current_row_col_value.iloc[0] +","+ current_row_col_value.iloc[1]+","+ current_row_col_value.iloc[2]+","+ current_row_col_value.iloc[3])
        print("8st")
    else : 
        print("ok")
            
#you can see the integrated row values in dataframe here please see after bunch of print statements 
df
```

```python
# here i am dropping the duplicate rows as previous operation resulted in duplicating the rows in dataframe

df = df.drop_duplicates()

# Check the transformed dataframe with no duplicate rows
df
```

```python
# those neighboorhoods colum rows which have not assigned in them i am replacing them with that rows Borough value
# it shows the row where transformation happened

for index, row in df.iterrows():
    if row['Neighbourhood'] == 'Not assigned':
        df.loc[index,'Neighbourhood'] = df.loc[index,'Borough']
        print("the change happened at row  " + index)
        

    

```
